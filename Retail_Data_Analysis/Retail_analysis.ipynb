Distribution of Data

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (20,20))
sns.distplot(df['Sales'],color = 'blue',ax = axs[0][0])
sns.distplot(df['Profit'],color = 'brown',ax = axs[0][1])
sns.distplot(df['Quantity'],color = 'red',ax = axs[1][0])
sns.distplot(df['Discount'],color = 'green',ax = axs[1][1])
axs[0][0].set_title('Sales Distribution',fontsize =30)
axs[0][1].set_title('Profit Distribution',fontsize =30)
axs[1][0].set_title('Quantity distribution',fontsize =30)
axs[1][1].set_title('Discount Distribution',fontsize =30)
plt.show()
```

Statewise Analysis

```python
df_state = df['State'].value_counts()
df_state.plot(kind='bar',figsize=(20,9),color="black")
plt.ylabel('Number of deals')
plt.xlabel('States')

plt.title('State Wise Dealings', fontsize = 20)
plt.show()
print("Average :")
df['State'].value_counts().mean()
```

Observation: The state with lowest deals is Wyoming
            The state with highest deals is California
            Average number of deal per state is 204.

Citywise Analysis

```python
df_city = df['City'].value_counts()
df_city=df_city.head(50)
df_city.plot(kind='bar',figsize=(20,10),color='red')
plt.ylabel('Number of deals')
plt.xlabel('City')
plt.title('City Wise Dealings', fontsize = 30)
plt.show()
print("Average :")
df['City'].value_counts().mean()
```

Observation: The city with highest deals is New York City.
            Average number of deal per city is 19.

Segmentwise Analysis

```python
df['Segment'].value_counts()
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
df_segment.plot.pie(subplots=True, 
                    autopct='%1.1f%%',
                    figsize=(20,20),
                    startangle=90,     
                    shadow=True,
                    labels = df_segment.index,
                   normalize=True)
plt.title(
    'Segmentwise analysis Parameters:\n' 
    'Sale, Discount, Profit',
    fontsize = 30
) 
```

Observation:-
Sales: Consumer-32%, Corporate - 33.5%, Home Office - 34.5%

Discount : Consumer-15.8%,  Corporate- 15.8% , Home Office-14.7%

Profit :Consumer-15.8%, Corporate- 15.8%, Home Office- 14.7%
